```python
# Checking outlier present or not
fig, axs = plt.subplots(2,3, figsize = (14,5))
sns.boxplot(df['Avg. Area Income'],ax=axs[0,0])
sns.boxplot(df['Avg. Area House Age'],ax=axs[0,1])
sns.boxplot(df['Area Population'],ax=axs[0,2])
sns.boxplot(df['Avg. Area Number of Rooms'],ax=axs[1,0])
sns.boxplot(df['Avg. Area Number of Bedrooms'],ax=axs[1,1])
sns.boxplot(df['Price'],ax=axs[1,2])
```

As we can see above that every columns as outlier except area numbr of bedrooms

```python
# outlier treatment for Avg. Area Income
plt.boxplot(df['Avg. Area Income'])
Q1 = df['Avg. Area Income'].quantile(0.25)
Q3 = df['Avg. Area Income'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Avg. Area Income'] >= Q1 - 1.5*IQR) & (df['Avg. Area Income'] <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment forAvg. Area House Age
plt.boxplot(df['Avg. Area House Age'])
Q1 = df['Avg. Area House Age'].quantile(0.25)
Q3 = df['Avg. Area House Age'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Avg. Area House Age'] >= Q1 - 1.5*IQR) & (df['Avg. Area House Age'] <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for Area Population
plt.boxplot(df['Area Population'])
Q1 = df['Area Population'].quantile(0.25)
Q3 = df['Area Population'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Area Population'] >= Q1 - 1.5*IQR) & (df['Area Population'] <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for Avg. Area Number of Rooms
plt.boxplot(df['Avg. Area Number of Rooms'])
Q1 = df['Avg. Area Number of Rooms'].quantile(0.25)
Q3 = df['Avg. Area Number of Rooms'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Avg. Area Number of Rooms'] >= Q1 - 1.5*IQR) & (df['Avg. Area Number of Rooms'] <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for Price
plt.boxplot(df['Price'])
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Price'] >= Q1 - 1.5*IQR) & (df['Price'] <= Q3 + 1.5*IQR)]
```

```python
# Resolve all outlier seeing the boxplot
fig, axs = plt.subplots(2,3, figsize = (18,5))
sns.boxplot(df['Avg. Area Income'],ax=axs[0,0])
sns.boxplot(df['Avg. Area House Age'],ax=axs[0,1])
sns.boxplot(df['Area Population'],ax=axs[0,2])
sns.boxplot(df['Avg. Area Number of Rooms'],ax=axs[1,0])
sns.boxplot(df['Avg. Area Number of Bedrooms'],ax=axs[1,1])
sns.boxplot(df['Price'],ax=axs[1,2])
```

```python
x = df.drop(['Address','Price'],axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(x_train)

# transform train and test sets
X_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
X_train_scaled
```

```python
x_train_scaled = pd.DataFrame(X_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(X_test_scaled, columns=x_test.columns)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score(y_test,y_pred)
```
